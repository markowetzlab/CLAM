```python
# import the csv
import csv
import pandas as pd

path = '/scratchc/fmlab/zuberi01/phd/CLAM/matching_rows.csv'
df = pd.read_csv(path)

# only keep the columns where the name of the column has a ':' in it
# and also the column 'slide_id'
# also the column 'label'
df = df[[col for col in df.columns if ':' in col] + ['slide_id'] + ['label']]

# print the number of columns and check that it's 1585-1024+2
print(len(df.columns))
print(len(df.columns) == 1585 - 1024 + 2)

# now put in a dictionary the slide_id as key and the list of columns as value and the label as well as value
data = {}
for index, row in df.iterrows():
    data[row['slide_id']] = list(row[:-2]) + [row['label']]

# now check that each slide_id has 561 columns
for key in data:
    if len(data[key]) != 562:
        print(key, len(data[key]))

# now print the first 5 and last 5 elements of the first 5 dictionaries
for key in list(data.keys())[:5] + list(data.keys())[-5:]:
    print(key, data[key][:5], data[key][-5:])
```

```python
import os

def process_csv_files(parent_folder):
    data_dicts = []

    # Iterate over the 10 CSV files
    for i in range(10):
        file_path = os.path.join(parent_folder, f'splits_{i}.csv')
        if os.path.exists(file_path):
            with open(file_path, mode='r') as file:
                csv_reader = csv.DictReader(file)
                
                # Initialise lists to store column data for the current file
                train_values = []
                val_values = []
                test_values = []
                
                for row in csv_reader:
                    train_values.append(row['train'])
                    val_values.append(row['val'])
                    test_values.append(row['test'])
                
                # Remove empty string values
                train_values = [value for value in train_values if value.strip()]
                val_values = [value for value in val_values if value.strip()]
                test_values = [value for value in test_values if value.strip()]
                
                # Create a dictionary for this split file
                split_dict = {
                    'train': train_values,
                    'val': val_values,
                    'test': test_values
                }
                
                data_dicts.append(split_dict)
        else:
            print(f"File not found: {file_path}")

    # Print the number of dictionaries in the list
    print(f"Number of dictionaries in the list: {len(data_dicts)}")
    return data_dicts

parent_folder = "/scratchc/fmlab/zuberi01/phd/splits_copy"
result = process_csv_files(parent_folder)

for i, d in enumerate(result): 
    print(f"Dictionary for split_{i}: {d}")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, roc_auc_score
import pandas as pd

# Placeholder for metrics for all models
all_metrics = []

# Process each split
for i in range(10):
    split_data = result[i]  # Assuming `result` stores the split data
    train_ids, val_ids, test_ids = split_data['train'], split_data['val'], split_data['test']

    # Prepare train data
    train_features = np.array([data[slide_id][:-1] for slide_id in train_ids if slide_id in data])
    train_labels = np.array([1 if data[slide_id][-1] == 'P' else 0 for slide_id in train_ids if slide_id in data])

    # Prepare validation data
    val_features = np.array([data[slide_id][:-1] for slide_id in val_ids if slide_id in data])
    val_labels = np.array([1 if data[slide_id][-1] == 'P' else 0 for slide_id in val_ids if slide_id in data])

    # Prepare test data
    test_features = np.array([data[slide_id][:-1] for slide_id in test_ids if slide_id in data])
    test_labels = np.array([1 if data[slide_id][-1] == 'P' else 0 for slide_id in test_ids if slide_id in data])

    # Train a Random Forest model
    clf = RandomForestClassifier(random_state=42, n_estimators=100)
    clf.fit(train_features, train_labels)

    # Get probabilities and predictions
    train_probs = clf.predict_proba(train_features)[:, 1]
    val_probs = clf.predict_proba(val_features)[:, 1]
    test_probs = clf.predict_proba(test_features)[:, 1]

    train_preds = (train_probs > 0.5).astype(int)
    val_preds = (val_probs > 0.5).astype(int)
    test_preds = (test_probs > 0.5).astype(int)

    # Calculate metrics
    train_acc = accuracy_score(train_labels, train_preds)
    val_acc = accuracy_score(val_labels, val_preds)
    test_acc = accuracy_score(test_labels, test_preds)

    train_recall = recall_score(train_labels, train_preds)
    val_recall = recall_score(val_labels, val_preds)
    test_recall = recall_score(test_labels, test_preds)

    train_auc = roc_auc_score(train_labels, train_probs)
    val_auc = roc_auc_score(val_labels, val_probs)
    test_auc = roc_auc_score(test_labels, test_probs)

    # Store metrics for this split
    all_metrics.append({
        'split': f'Split {i}',
        'train_accuracy': train_acc,
        'val_accuracy': val_acc,
        'test_accuracy': test_acc,
        'train_recall': train_recall,
        'val_recall': val_recall,
        'test_recall': test_recall,
        'train_auc': train_auc,
        'val_auc': val_auc,
        'test_auc': test_auc
    })

# Create a DataFrame for all metrics
metrics_df = pd.DataFrame(all_metrics)

# Add averages to the table
averages = pd.DataFrame([{
    'split': 'Average',  # Explicitly set the split column for averages
    'train_accuracy': metrics_df['train_accuracy'].mean(),
    'val_accuracy': metrics_df['val_accuracy'].mean(),
    'test_accuracy': metrics_df['test_accuracy'].mean(),
    'train_recall': metrics_df['train_recall'].mean(),
    'val_recall': metrics_df['val_recall'].mean(),
    'test_recall': metrics_df['test_recall'].mean(),
    'train_auc': metrics_df['train_auc'].mean(),
    'val_auc': metrics_df['val_auc'].mean(),
    'test_auc': metrics_df['test_auc'].mean()
}])

metrics_df = pd.concat([metrics_df, averages], ignore_index=True)

# Save the DataFrame to a CSV file for review
output_path = "model_metrics_summary.csv"
metrics_df.to_csv(output_path, index=False)

# Display the final table in the console
print(metrics_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, ElasticNet
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score

# List of models to evaluate
models = {
    'RandomForest': RandomForestClassifier(random_state=42, n_estimators=100),
    'LogisticRegression': LogisticRegression(random_state=42, max_iter=1000),
    'SVM': SVC(probability=True, random_state=42),  # Enable probabilities for AUC
    'KNN': KNeighborsClassifier(),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'ElasticNetRegression': ElasticNet(random_state=42, alpha=1.0, l1_ratio=0.5)
}

# Placeholder for averaged metrics for each model
model_metrics = []

# Loop through each model
for model_name, model in models.items():
    all_metrics = []  # Placeholder for metrics of all splits for the current model

    # Process each split
    for i in range(10):
        split_data = result[i]  # Assuming `result` stores the split data
        train_ids, val_ids, test_ids = split_data['train'], split_data['val'], split_data['test']

        # Prepare train data
        train_features = np.array([data[slide_id][:-1] for slide_id in train_ids if slide_id in data])
        train_labels = np.array([1 if data[slide_id][-1] == 'P' else 0 for slide_id in train_ids if slide_id in data])

        # Prepare validation data
        val_features = np.array([data[slide_id][:-1] for slide_id in val_ids if slide_id in data])
        val_labels = np.array([1 if data[slide_id][-1] == 'P' else 0 for slide_id in val_ids if slide_id in data])

        # Prepare test data
        test_features = np.array([data[slide_id][:-1] for slide_id in test_ids if slide_id in data])
        test_labels = np.array([1 if data[slide_id][-1] == 'P' else 0 for slide_id in test_ids if slide_id in data])

        # Train the model
        if model_name == 'ElasticNetRegression':
            from sklearn.preprocessing import StandardScaler
            scaler = StandardScaler()
            
            # Scale features
            train_features = scaler.fit_transform(train_features)
            val_features = scaler.transform(val_features)
            test_features = scaler.transform(test_features)

            # Use LogisticRegression with ElasticNet penalty for classification
            elasticnet_model = LogisticRegression(penalty='elasticnet', solver='saga', l1_ratio=0.5, random_state=42, max_iter=1000)
            elasticnet_model.fit(train_features, train_labels)
            
            train_probs = elasticnet_model.predict_proba(train_features)[:, 1]
            val_probs = elasticnet_model.predict_proba(val_features)[:, 1]
            test_probs = elasticnet_model.predict_proba(test_features)[:, 1]

            # Convert probabilities to binary predictions
            train_preds = (train_probs > 0.5).astype(int)
            val_preds = (val_probs > 0.5).astype(int)
            test_preds = (test_probs > 0.5).astype(int)

        else:
            model.fit(train_features, train_labels)
            train_probs = model.predict_proba(train_features)[:, 1]
            val_probs = model.predict_proba(val_features)[:, 1]
            test_probs = model.predict_proba(test_features)[:, 1]

        # Convert probabilities to binary predictions
        train_preds = (train_probs > 0.5).astype(int)
        val_preds = (val_probs > 0.5).astype(int)
        test_preds = (test_probs > 0.5).astype(int)

        # Calculate metrics
        train_acc = accuracy_score(train_labels, train_preds)
        val_acc = accuracy_score(val_labels, val_preds)
        test_acc = accuracy_score(test_labels, test_preds)

        train_recall = recall_score(train_labels, train_preds)
        val_recall = recall_score(val_labels, val_preds)
        test_recall = recall_score(test_labels, test_preds)

        train_auc = roc_auc_score(train_labels, train_probs)
        val_auc = roc_auc_score(val_labels, val_probs)
        test_auc = roc_auc_score(test_labels, test_probs)

        # Store metrics for this split
        all_metrics.append({
            'train_accuracy': train_acc,
            'val_accuracy': val_acc,
            'test_accuracy': test_acc,
            'train_recall': train_recall,
            'val_recall': val_recall,
            'test_recall': test_recall,
            'train_auc': train_auc,
            'val_auc': val_auc,
            'test_auc': test_auc
        })

    # Calculate averaged metrics for the current model
    avg_metrics = pd.DataFrame(all_metrics).mean().to_dict()
    avg_metrics['model'] = model_name  # Add the model name
    model_metrics.append(avg_metrics)

# Create a DataFrame for all models' averaged metrics
final_metrics_df = pd.DataFrame(model_metrics)

# Save the summary table to a CSV file for review
output_path = "model_averaged_metrics_summary.csv"
final_metrics_df.to_csv(output_path, index=False)

# Display the final table in the console
print(final_metrics_df)
```
